```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

```python
data = load_boston()     #import dataframe
X = pd.DataFrame(data["data"], columns=data["feature_names"])    #X là các cột features
y = pd.DataFrame(data["target"], columns=["MEDV"])   #Y la cột giá nhà
df = pd.concat([X, y], axis=1)   #ghép X và Y với nhau theo chiều ngang
df_train, df_test = train_test_split(df, test_size=0.1, random_state=1)   #chia dataset thành 2 tập train và test
```

```python
print(df_train.shape)
df_train.head()
```

```python
model_cart = DecisionTreeRegressor(criterion="mse", max_depth=3, min_samples_split=20)  #khởi tạo model
                                                                                        #criterion: sử dụng hàm loss MSE
                                                                                        #max depth là độ sâu tối đa của cây quyết định
                                                                                        #min_samples_split: mỗi lá có ít nhất 20 data samples data samples để phân tách left và right
```

```python
model_cart.fit(df_train.drop("MEDV", axis=1), df_train["MEDV"]) #fit các cột features và cột Y vào model
```

```python
df_train
```

```python
y_pred = model_cart.predict(df_test.drop("MEDV", axis=1))   #predict các giá trị của tập test
```

```python
mean_squared_error(y_pred, df_test["MEDV"])    #tính MSE theo mô hình CART
```

```python
model_rf = RandomForestRegressor(criterion="mse",     #sử dụng hàm loss MSE
                              max_depth=3,            #độ sâu tối đa của cây quyết định là 3
                              min_samples_split=20,   #mỗi lá có ít nhất 20 data samples để phân tách left và right
                              n_estimators=10)        #rừng có khoảng 10 cây
model_rf.fit(df_train.drop("MEDV", axis=1), df_train["MEDV"]) #fit dữ liệu train vào model
y_pred = model_rf.predict(df_test.drop("MEDV", axis=1))  
```

```python
mean_squared_error(y_pred, df_test["MEDV"]) #tính MSE theo mô hình RF
```

```python
model_rf.estimators_[0]
```
