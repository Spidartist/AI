```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(-10,10,0.1)
y = x**2
plt.plot(x,y)
plt.title("$y = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = np.arange(-10,10,0.1)
plt.figure(figsize=(5,30))
y1 = x**2
y2 = x**3
plt.plot(x,y1,label="$y=x^2$")
plt.plot(x,y2,label="$y=x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
